#  A loss curve that shows the loss in the y-axis, and training iterations in the x-axis





### Cross-entropy loss is a general measure of how well predicted class probabilities compare to their respective actual classes on a logarithmic scale. When applied to the loss function, the difference between the two probability distributions outputs a loss value, which, when reviewed over multiple iterations, shows the likelihood of accurate predictions by the model. A cross-entropy loss value of 0 would indicate a perfectly trained model. The trendline in the graph above provides a negative correlation which implies that the model is becoming more accurate in its predictions.
